```python
import math
from typing import Dict, Union, Optional

def circle_calculator(
    radius: Optional[float] = None,
    diameter: Optional[float] = None,
    area: Optional[float] = None,
    circumference: Optional[float] = None
) -> Dict[str, Union[float, str]]:
    """
    Calculate properties of a circle based on one given parameter.
    
    Args:
        radius: The radius of the circle
        diameter: The diameter of the circle
        area: The area of the circle
        circumference: The circumference of the circle
    
    Returns:
        Dictionary containing all properties of the circle or an error message
    """
    # Determine which parameter was provided
    if radius is not None:
        if radius <= 0:
            return {"error": "Radius must be positive"}
        r = radius
    elif diameter is not None:
        if diameter <= 0:
            return {"error": "Diameter must be positive"}
        r = diameter / 2
    elif area is not None:
        if area <= 0:
            return {"error": "Area must be positive"}
        r = (area / math.pi) ** 0.5  # More efficient than math.sqrt
    elif circumference is not None:
        if circumference <= 0:
            return {"error": "Circumference must be positive"}
        r = circumference / (2 * math.pi)
    else:
        return {"error": "No parameters provided"}
    
    # Pre-calculate values used multiple times
    pi_r = math.pi * r
    
    # Return all properties at once
    return {
        "radius": r,
        "diameter": 2 * r,
        "area": pi_r * r,
        "circumference": 2 * pi_r
    }

# Example usage
if __name__ == "__main__":
    # Test with different inputs
    test_cases = [
        {"radius": 5},
        {"diameter": 10},
        {"area": 100},
        {"circumference": 30},
        {"radius": -5},  # Error case
        {}  # Error case
    ]
    
    for case in test_cases:
        result = circle_calculator(**case)
        
        if "error" in result:
            print(f"\nInput {case}: {result['error']}")
        else:
            print(f"\nInput {case}:")
            for key, value in result.items():
                print(f"{key}: {value:.4f}")
```

```python
def create_user_profile(name, age, occupation="Student", interests=None): # Use None as default
    """
    Creates a user profile with optional interests.

    Args:
        name (str): The user's name (required).
        age (int): The user's age (required).
        occupation (str, optional): The user's occupation (defaults to "Student").
        interests (list, optional): A list of the user's interests (defaults to None).
    """
    if interests is None:  # Initialize if None
        interests = [] 

    profile = {
        "name": name,
        "age": age,
        "occupation": occupation,
        "interests": interests
    }

    return profile

# Usage
user1 = create_user_profile("Alice", 25, "Software Engineer", ["Coding", "Hiking"])
user2 = create_user_profile("Bob", 18)  # Uses default occupation and no interests
user3 = create_user_profile("Carol", 30, interests=["Gardening", "Reading"])

print(user1)
print(user2)
print(user3)
```

### Binary Search [Lists]

```python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid  # devuelve la posición del producto
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1  # si no se encuentra el producto

# Lista de precios de productos (ordenada)
precios = [5, 12, 20, 23, 30, 35, 45, 50, 60, 75]

# Buscamos el producto que cuesta 45
precio_buscar = 45
index = binary_search(precios, precio_buscar)

if index != -1:
    print(f"Producto encontrado! Precio: ${precios[index]}, Posición en la lista: {index}")
else:
    print(f"Producto con precio ${precio_buscar} no encontrado")
```
