````python
import math
from typing import Dict, Union, Optional

def circle_calculator(
    radius: Optional[float] = None,
    diameter: Optional[float] = None,
    area: Optional[float] = None,
    circumference: Optional[float] = None
) -> Dict[str, Union[float, str]]:
    """
    Calculate properties of a circle based on one given parameter.
    
    Args:
        radius: The radius of the circle
        diameter: The diameter of the circle
        area: The area of the circle
        circumference: The circumference of the circle
    
    Returns:
        Dictionary containing all properties of the circle or an error message
    """
    # Determine which parameter was provided
    if radius is not None:
        if radius <= 0:
            return {"error": "Radius must be positive"}
        r = radius
    elif diameter is not None:
        if diameter <= 0:
            return {"error": "Diameter must be positive"}
        r = diameter / 2
    elif area is not None:
        if area <= 0:
            return {"error": "Area must be positive"}
        r = (area / math.pi) ** 0.5  # More efficient than math.sqrt
    elif circumference is not None:
        if circumference <= 0:
            return {"error": "Circumference must be positive"}
        r = circumference / (2 * math.pi)
    else:
        return {"error": "No parameters provided"}
    
    # Pre-calculate values used multiple times
    pi_r = math.pi * r
    
    # Return all properties at once
    return {
        "radius": r,
        "diameter": 2 * r,
        "area": pi_r * r,
        "circumference": 2 * pi_r
    }

# Example usage
if __name__ == "__main__":
    # Test with different inputs
    test_cases = [
        {"radius": 5},
        {"diameter": 10},
        {"area": 100},
        {"circumference": 30},
        {"radius": -5},  # Error case
        {}  # Error case
    ]
    
    for case in test_cases:
        result = circle_calculator(**case)
        
        if "error" in result:
            print(f"\nInput {case}: {result['error']}")
        else:
            print(f"\nInput {case}:")
            for key, value in result.items():
                print(f"{key}: {value:.4f}")
````

### Binary Search [Lists]

````python
def binary_search(arr, target):
    left = 0
    right = len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid  # devuelve la posición del producto
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1  # si no se encuentra el producto

# Lista de precios de productos (ordenada)
precios = [5, 12, 20, 23, 30, 35, 45, 50, 60, 75]

# Buscamos el producto que cuesta 45
precio_buscar = 45
index = binary_search(precios, precio_buscar)

if index != -1:
    print(f"Producto encontrado! Precio: ${precios[index]}, Posición en la lista: {index}")
else:
    print(f"Producto con precio ${precio_buscar} no encontrado")
````

````python
# Function to perform full dataset profiling
# ----------------------------
def profile_dataset(df, cat_threshold=10):
    """
    Perform a comprehensive EDA and dataset profiling.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    cat_threshold (int): Max unique values for numerical columns to be treated as categorical.

    Returns:
    pd.DataFrame: Summary table with column info, statistics, and missing values.
    """
    
    # Identify column data types
    column_types = df.dtypes.to_dict()
    
    # Separate numerical and categorical columns
    numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_cols = df.select_dtypes(include=['object', 'category', 'bool']).columns.tolist()
    
    # Treat numerical columns with few unique values as categorical
    numerical_but_categorical = [col for col in numerical_cols if df[col].nunique() <= cat_threshold]
    true_numerical_cols = [col for col in numerical_cols if col not in numerical_but_categorical]
    
    categorical_cols += numerical_but_categorical  # Add numerical-but-categorical
    categorical_cols = list(set(categorical_cols))  # Remove duplicates
    
    # Build summary table
    summary_list = []
    
    for col in df.columns:
        dtype = df[col].dtype
        
        # Determine variable type
        if col in true_numerical_cols:
            var_type = "Numerical"
        else:
            var_type = "Categorical"
        
        # Count unique values
        unique_vals = df[col].nunique()
        
        # Count missing values
        null_count = df[col].isnull().sum()
        not_null_count = df[col].notnull().sum()
        
        # Compute numerical statistics if applicable
        if var_type == "Numerical":
            mean = df[col].mean()
            median = df[col].median()
            std = df[col].std()
            min_val = df[col].min()
            max_val = df[col].max()
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            deviation = std / mean if mean != 0 else np.nan
        else:
            mean = median = std = min_val = max_val = q1 = q3 = deviation = np.nan
        
        # Top category info for categorical variables
        if var_type == "Categorical":
            top_category = df[col].mode()[0] if not df[col].mode().empty else np.nan
            top_freq = df[col].value_counts().iloc[0] if not df[col].value_counts().empty else np.nan
        else:
            top_category = top_freq = np.nan
        
        # Check for constant or highly imbalanced columns
        constant = True if unique_vals == 1 else False
        highly_imbalanced = True if var_type == "Categorical" and (top_freq / not_null_count > 0.9) else False
        
        summary_list.append({
            "Column": col,
            "Dtype": dtype,
            "Variable Type": var_type,
            "Unique Values": unique_vals,
            "Null Values": null_count,
            "Not Null Values": not_null_count,
            "Mean": mean,
            "Median": median,
            "Std Dev": std,
            "Min": min_val,
            "Max": max_val,
            "Q1": q1,
            "Q3": q3,
            "Deviation": deviation,
            "Top Category": top_category,
            "Top Frequency": top_freq,
            "Constant": constant,
            "Highly Imbalanced": highly_imbalanced
        })
    
    summary_df = pd.DataFrame(summary_list)
    
    return summary_df


# Example usage
# ----------------------------

summary = profile_dataset(df)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(summary)
````

````python
# ==========================================
# # Chapter 1: Import Libraries
# ==========================================
import numpy as np
import pandas as pd

# ==========================================
# # Chapter 2: Create a Dataset
# ==========================================
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 40, 28, 26, 31, 36, 41],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "New York", "Los Angeles", "Chicago", "Houston"],
    "Salary": [70000, 80000, 120000, 90000, 75000, 72000, 82000, 125000, 91000],
    "Department": ["HR", "IT", "IT", "Finance", "HR", "HR", "IT", "IT", "Finance"],
    "Bonus": [5000, 7000, 10000, 8000, 4000, 5200, 7300, 10500, 8500]
}

df_Example = pd.DataFrame(data)
print("# Dataset:")
print(df_Example)

# ==========================================
# # Chapter 3: Advanced Pivot Table
# ==========================================
# Pivot table: Aggregate Salary and Bonus by Department and City
pivot_advanced = pd.pivot_table(
    df_Example,
    values=["Salary", "Bonus"],           # Multiple values to aggregate
    index=["Department", "City"],         # Multi-level row index
    aggfunc={"Salary": [np.mean, np.sum], # Multiple aggregation functions
             "Bonus": [np.sum, np.max]},
    fill_value=0                           # Replace NaN with 0
)

print("\n# Advanced Pivot Table:")
print(pivot_advanced)

# ==========================================
# # Chapter 4: Flatten MultiIndex Columns
# ==========================================
# Flatten multi-level columns for easier access
pivot_advanced.columns = ['_'.join(col).strip() for col in pivot_advanced.columns.values]
pivot_flat = pivot_advanced.reset_index()

print("\n# Flattened Advanced Pivot Table:")
print(pivot_flat)

# ==========================================
# # Chapter 5: Sorting and Filtering Pivot
# ==========================================
# Sort by total Salary sum descending
pivot_sorted = pivot_flat.sort_values(by="Salary_sum", ascending=False)
print("\n# Sorted Pivot Table by Total Salary:")
print(pivot_sorted)

# Filter departments where mean Salary > 80000
high_salary = pivot_flat[pivot_flat["Salary_mean"] > 80000]
print("\n# Departments with Mean Salary > 80000:")
print(high_salary)
````

#### Inconsistences

````python
import pandas as pd
import numpy as np
from scipy import stats

def detect_inconsistencies(df):
    report = []

    for col in df.columns:
        col_data = df[col]
        issues = []
        details = {}

        # Missing values
        missing_pct = col_data.isna().mean() * 100
        if missing_pct > 30:
            issues.append("High missing values")
            details["missing_pct"] = round(missing_pct, 2)

        # Mixed data types
        unique_types = col_data.dropna().map(type).nunique()
        if unique_types > 1:
            issues.append("Mixed data types")
            details["types"] = col_data.dropna().map(type).unique().astype(str).tolist()

        # High cardinality
        if col_data.nunique() > 0.9 * len(df):
            issues.append("Very high cardinality")
            details["unique_count"] = col_data.nunique()

        # Categorical noise (rare categories)
        if col_data.dtype == 'object' or col_data.dtype.name == 'category':
            freq = col_data.value_counts(normalize=True)
            rare = freq[freq < 0.01]
            if len(rare) > 0:
                issues.append("Rare categories")
                details["rare_values"] = rare.index.tolist()

        # Numerical anomalies
        if pd.api.types.is_numeric_dtype(col_data):
            # Negative values
            if (col_data < 0).any():
                issues.append("Negative values present")
                details["neg_count"] = int((col_data < 0).sum())
            # Outliers (z-score)
            z_scores = np.abs(stats.zscore(col_data.dropna()))
            outlier_count = (z_scores > 3).sum()
            if outlier_count > 0:
                issues.append("Extreme outliers")
                details["outlier_count"] = int(outlier_count)

        if issues:
            report.append({
                "column": col,
                "issues": ", ".join(issues),
                "details": details
            })

    return pd.DataFrame(report)

# Example usage:
summary = detect_inconsistencies(df)
display(summary)
````

#### Caso Práctico - Fuga de clientes

Considerando la importancia actual de las telecomunicaciones en el desarrollo de las actividades humanas, los clientes de las empresas de telecomunicaciones buscan un servicio eficiente que soporte las diferentes actividades que realizan, con llamadas, datos, servicio de internet, mensajes de texto, y servicio al cliente. Al contratar un paquete de telecomunicaciones con una empresa, los clientes esperan obtener una buena relación costo/beneficio, de tal forma que el servicio que recibe el cliente compense el gasto total mensual del mismo. Así las cosas, si los clientes sufren ciertas fallas en el servicio o en la calidad de la atención que reciben, se ven incentivados a cancelar la suscripción y contratar servicios de telecomunicaciones con otra empresa. El archivo de datos Fuga de Clientes, recopila la información de 289 clientes y el uso que cada uno de ellos hace de su plan de telecomunicaciones. Las variables que contiene el archivo de datos se encuentran descritas a continuación:

Variables 

ID Cliente: número de identificación del cliente en la red.

Edad en la red: tiempo que ha pasado, en meses, desde que el cliente ingresó a la red.

Gasto total al mes: gasto total del cliente en la red, en el mes de agosto de 2017.

Gasto en SMS por SMS: el gasto total en SMS (Short Message Service) por cliente obtenido a través del servicio de SMS utilizado por el suscriptor.

Gasto de datos por SMS: gasto total de datos/Internet de un cliente obtenido a través del servicio de SMS utilizado por el suscriptor. 

Consumo total de datos: cantidad de datos/Internet, en KB, consumida por el cliente durante el mes de agosto de 2017.

Velocidad promedio de carga: velocidad promedio de carga (tiempo que tarda la información en llegar desde el celular a internet), en megabytes por segundo MB/S.  

Velocidad promedio de descarga: velocidad promedio de descarga (tiempo que tarda la información en llegar al celular desde internet), en megabytes por segundo MB/S.   

Llamadas únicas: número total de llamadas únicas realizadas durante el mes de agosto de 2017. 

Llamadas caídas: número total de llamadas caídas durante el mes de agosto de 2017.

Duración promedio llamada: duración promedio, en segundos, de las llamadas realizadas durante el mes de agosto de 2017.

Duración promedio de las llamadas al mismo operador: duración promedio, en segundos, de las llamadas realizadas al mismo operador durante el mes de agosto de 2017. 

Duración promedio de las llamadas a otros operadores: duración promedio, en segundos, de las llamadas realizadas a operadores diferentes al del cliente durante el mes de agosto de 2017.  

Gasto llamadas mismo operador: gasto total en llamadas del mismo operador.

Gasto llamadas otro operador: gasto total del cliente en llamadas realizadas a otros operadores.

Llamadas por quejas al mes: número de llamadas por quejas realizadas por el cliente durante el mes de agosto de 2017.

Tipo de red preferida: tipo de red preferida por el cliente.

Competidor favorito: compañía favorita a la cual se iría el cliente.

Cancelación de la suscripción: 1, si el cliente se fue de la red (cambió de operador), 0 de lo contrario.

#### Campana de gauss


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Definir los parámetros de la distribución
# Usaremos como ejemplo el Coeficiente Intelectual (IQ)
media = 100            # El centro de la campana
desviacion_std = 15    # Qué tan "dispersos" están los datos

# 2. Crear el rango de valores para el eje X
# Generamos valores desde -4 hasta +4 desviaciones estándar del centro
x = np.linspace(media - 4 * desviacion_std, media + 4 * desviacion_std, 1000)

# 3. Calcular la "línea" de la campana (Densidad de Probabilidad)
y = norm.pdf(x, media, desviacion_std)

# 4. Configurar y crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'Distribución Normal\n(μ={media}, σ={desviacion_std})', color='blue', lw=2)

# 5. Opcional: Sombrear el área del 68% (una desviación estándar)
x_68 = np.linspace(media - desviacion_std, media + desviacion_std, 100)
plt.fill_between(x_68, norm.pdf(x_68, media, desviacion_std), color='skyblue', alpha=0.5, label='68.2% central')

# Personalización de etiquetas
plt.title('Ejemplo de Campana de Gauss (Distribución Normal)', fontsize=14)
plt.xlabel('Puntaje / Valor', fontsize=12)
plt.ylabel('Probabilidad', fontsize=12)
plt.axvline(media, color='red', linestyle='--', label='Media') # Línea central
plt.legend()
plt.grid(True, alpha=0.3)

# Mostrar el gráfico
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom

# Configuración inicial
p = 0.5  # Probabilidad de éxito (50%)
n_binomial = 10  # Intentos para la Binomial

# 1. Bernoulli (Un solo intento)
x_bern = [0, 1]
y_bern = bernoulli.pmf(x_bern, p)

# 2. Binomial (Éxitos en 10 intentos)
x_bin = np.arange(0, n_binomial + 1)
y_bin = binom.pmf(x_bin, n_binomial, p)

# 3. Geométrica (Intentos hasta el primer éxito)
x_geom = np.arange(1, 11)
y_geom = geom.pmf(x_geom, p)

# Graficar
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# Gráfico Bernoulli
ax[0].bar(x_bern, y_bern, color='skyblue', edgecolor='black')
ax[0].set_title('Bernoulli (n=1)\n¿Cara o Cruz?')
ax[0].set_xticks([0, 1])
ax[0].set_xticklabels(['Fracaso', 'Éxito'])

# Gráfico Binomial
ax[1].bar(x_bin, y_bin, color='salmon', edgecolor='black')
ax[1].set_title('Binomial (n=10)\n¿Cuántas caras en total?')
ax[1].set_xlabel('Número de éxitos')

# Gráfico Geométrica
ax[2].bar(x_geom, y_geom, color='lightgreen', edgecolor='black')
ax[2].set_title('Geométrica\n¿En qué intento sale la 1ra cara?')
ax[2].set_xlabel('Número de intento')

plt.tight_layout()
plt.show()
````

````python
from scipy.stats import poisson

# Parámetros
l_3 = 3
l_10 = 10

# 1. Probabilidad de 0 visitas
p0_3 = poisson.pmf(0, l_3)
p0_10 = poisson.pmf(0, l_10)

# 2. Valor más probable (Moda)
# En Poisson, la moda es floor(lambda) y ceil(lambda)-1 si lambda es entero
# Para simplificar, buscaremos el valor con PMF máxima en un rango
import numpy as np
x = np.arange(0, 30)
pmf_3 = poisson.pmf(x, l_3)
pmf_10 = poisson.pmf(x, l_10)
moda_3 = x[np.argmax(pmf_3)]
moda_10 = x[np.argmax(pmf_10)]

# 3. Probabilidad de recibir MÁS que el promedio
# P(X > lambda) = 1 - P(X <= lambda)
p_gt_3 = 1 - poisson.cdf(l_3, l_3)
p_gt_10 = 1 - poisson.cdf(l_10, l_10)

# 4. Probabilidad de un rango cercano a la media (media +/- 1 desv. std)
# Desv Std = sqrt(lambda)
std_3 = l_3**0.5
std_10 = l_10**0.5
range_3 = (int(l_3 - std_3), int(l_3 + std_3))
prob_range_3 = poisson.cdf(range_3[1], l_3) - poisson.cdf(range_3[0] - 1, l_3)
range_10 = (int(l_10 - std_10), int(l_10 + std_10))
prob_range_10 = poisson.cdf(range_10[1], l_10) - poisson.cdf(range_10[0] - 1, l_10)

print(f"Lambda 3 - Prob(0): {p0_3:.4f}")
print(f"Lambda 3 - Moda: {moda_3}")
print(f"Lambda 3 - Prob(>3): {p_gt_3:.4f}")
print(f"Lambda 3 - Prob en rango {range_3}: {prob_range_3:.4f}")

print(f"Lambda 10 - Prob(0): {p0_10:.4f}")
print(f"Lambda 10 - Moda: {moda_10}")
print(f"Lambda 10 - Prob(>10): {p_gt_10:.4f}")
print(f"Lambda 10 - Prob en rango {range_10}: {prob_range_10:.4f}")
````

#### Variables aleatorias continua y Funcion de densidad


````python
##### Para que este ejemplo sea realmente útil, no solo veremos la gráfica, sino que calcularemos datos reales. Vamos a suponer 
# que somos analistas de datos para una tienda de ropa y queremos saber cuántas personas entran en el rango de "talla mediana".#
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- 1. CONFIGURACIÓN DEL CASO ---
# Supongamos que la estatura de una población sigue una distribución normal
media_estatura = 170  # Promedio de 170 cm
desv_estandar = 7     # La mayoría varía +/- 7cm respecto al promedio

# --- 2. CÁLCULO DE PROBABILIDAD (INSIGHT) ---
# Queremos saber: ¿Qué porcentaje de la población mide entre 165cm y 175cm?
limite_inferior = 165
limite_superior = 175

# La probabilidad es el área bajo la curva (se calcula restando las acumuladas)
probabilidad = norm.cdf(limite_superior, media_estatura, desv_estandar) - \
               norm.cdf(limite_inferior, media_estatura, desv_estandar)

# --- 3. GENERACIÓN DE LA GRÁFICA ---
x = np.linspace(140, 200, 1000)
pdf = norm.pdf(x, media_estatura, desv_estandar)

plt.figure(figsize=(12, 6))
plt.plot(x, pdf, label='Curva de Densidad (Población)', color='#2c3e50', lw=3)

# Sombrear el rango de interés
x_insight = np.linspace(limite_inferior, limite_superior, 100)
plt.fill_between(x_insight, norm.pdf(x_insight, media_estatura, desv_estandar), 
                 color='#3498db', alpha=0.4, label='Rango Talla M')

# Anotaciones de Insights
plt.title('Análisis de Mercado: Distribución de Estaturas', fontsize=15)
plt.xlabel('Estatura (cm)', fontsize=12)
plt.ylabel('Densidad de Probabilidad', fontsize=12)
plt.axvline(media_estatura, color='red', linestyle='--', label=f'Media ({media_estatura}cm)')

# Mostrar el resultado del cálculo en la gráfica
plt.text(145, 0.04, f"INSIGHT:\nEl {probabilidad*100:.2f}% de las personas\n"
                   f"miden entre {limite_inferior} y {limite_superior} cm", 
         bbox=dict(facecolor='white', alpha=0.8), fontsize=11)

plt.legend()
plt.grid(alpha=0.3)
plt.show()

print(f"Resultado: La probabilidad de encontrar a alguien en ese rango es de {probabilidad:.4f}")
````

#### Distribuciones uniforme, normal y exponencial

````python
#1. Distribución Uniforme: El tiempo de espera del Metro
#Este código simula la llegada aleatoria de un usuario a una estación donde el tren pasa exactamente cada 10 minutos.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Rango: de 0 a 10 minutos
inicio = 0
ancho = 10

# Generar datos para la curva
x = np.linspace(-2, 12, 1000)
pdf = uniform.pdf(x, loc=inicio, scale=ancho)

plt.figure(figsize=(10, 4))
plt.plot(x, pdf, color='green', lw=3, label='Densidad Uniforme')
plt.fill_between(x, pdf, alpha=0.2, color='green')

# Insight: Probabilidad de esperar entre 2 y 5 minutos
p_espera = uniform.cdf(5, 0, 10) - uniform.cdf(2, 0, 10)

plt.title('Distribución Uniforme: Tiempo de espera del Metro')
plt.xlabel('Minutos')
plt.ylabel('Densidad')
plt.text(0, 0.05, f"INSIGHT:\nLa probabilidad de esperar\nentre 2 y 5 min es del {p_espera*100:.0f}%", bbox=dict(facecolor='white'))
plt.grid(alpha=0.3)
plt.show()

#Insights de la Distribución Uniforme:

#Equiprobabilidad: No hay un "pico". La probabilidad de esperar el primer minuto es la misma que la del último.

#Predicción: Es ideal para procesos donde no hay sesgos, como la generación de números aleatorios para sorteos o tiempos de espera en ciclos constantes.
````

````python
#2. Distribución Normal: Puntajes de un Examen Estándar
#Aquí simulamos los resultados de un examen donde la mayoría de los estudiantes sacan una nota cercana al promedio (ej. 70/100).

from scipy.stats import norm

media_nota = 70
desv_est = 10

x = np.linspace(30, 110, 1000)
pdf = norm.pdf(x, media_nota, desv_est)

plt.figure(figsize=(10, 4))
plt.plot(x, pdf, color='blue', lw=3, label='Densidad Normal')

# Insight: Regla Empírica (Área entre 60 y 80)
plt.fill_between(x, pdf, where=(x >= 60) & (x <= 80), alpha=0.4, color='blue')

plt.title('Distribución Normal: Puntajes de Examen')
plt.xlabel('Puntaje')
plt.text(35, 0.02, "INSIGHT:\nCasi el 68% de los alumnos\nsacaron entre 60 y 80 puntos", bbox=dict(facecolor='white'))
plt.grid(alpha=0.3)
plt.show()

#Tendencia Central: La mayoría de los datos se agrupan en el centro. Los valores extremos (muy inteligentes o muy bajas notas) son muy raros.

#Simetría: La probabilidad de sacar 10 puntos por encima de la media es la misma que 10 puntos por debajo.
````

````python
#3. Distribución Exponencial: Tiempo entre llamadas en un Call Center
#Esta distribución es clásica para medir el tiempo entre eventos. Si las llamadas llegan rápido, el tiempo de espera entre una y otra será corto la mayoría de las veces.

from scipy.stats import expon

# Escala (1 / lambda): si llegan 2 llamadas por minuto, la escala es 0.5
tiempo_promedio_entre_llamadas = 0.5 

x = np.linspace(0, 4, 1000)
pdf = expon.pdf(x, scale=tiempo_promedio_entre_llamadas)

plt.figure(figsize=(10, 4))
plt.plot(x, pdf, color='red', lw=3, label='Densidad Exponencial')
plt.fill_between(x, pdf, alpha=0.2, color='red')

# Insight: Probabilidad de que la siguiente llamada sea en menos de 1 minuto
prob_rapida = expon.cdf(1, scale=tiempo_promedio_entre_llamadas)

plt.title('Distribución Exponencial: Tiempo entre Llamadas')
plt.xlabel('Minutos entre eventos')
plt.text(1.5, 1.0, f"INSIGHT:\nHay un {prob_rapida*100:.1f}% de probabilidad\nde que la próxima llamada entre\nen menos de 1 minuto.", bbox=dict(facecolor='white'))
plt.grid(alpha=0.3)
plt.show()

#Insights de la Distribución Exponencial:

#Falta de Memoria: A la distribución no le importa cuánto tiempo ha pasado ya; la probabilidad de que ocurra el evento en el siguiente segundo sigue siendo la misma.
#Decaimiento: Es muy probable que los eventos ocurran con poco tiempo de diferencia, y se vuelve "exponencialmente" menos probable que pase mucho tiempo sin que nada ocurra.
````

````python
# Ejemplo distribucion exponencial ecommerce
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# 1. PARÁMETROS DEL CASO
# El tiempo promedio (escala) de compra es de 20 minutos
promedio_minutos = 20
x = np.linspace(0, 120, 1000)  # Analizamos las primeras 2 horas
pdf = expon.pdf(x, scale=promedio_minutos)

# 2. CÁLCULO DE PROBABILIDAD (Insight)
# ¿Qué porcentaje de las ventas ocurren en los primeros 30 minutos?
limite_tiempo = 30
prob_30_min = expon.cdf(limite_tiempo, scale=promedio_minutos)

# 3. GRAFICAR
plt.figure(figsize=(12, 6))
plt.plot(x, pdf, color='#e67e22', lw=3, label='Probabilidad de Compra')

# Sombrear la ventana crítica (0 a 30 min)
x_ventana = np.linspace(0, limite_tiempo, 100)
plt.fill_between(x_ventana, expon.pdf(x_ventana, scale=promedio_minutos), 
                 color='#f1c40f', alpha=0.5, label='Ventana Crítica (0-30 min)')

plt.title('E-commerce: Tiempo desde Clic hasta Compra (Distribución Exponencial)')
plt.xlabel('Minutos después del Clic')
plt.ylabel('Densidad de Probabilidad')
plt.grid(alpha=0.2)
plt.legend()
plt.savefig('ecommerce_exponential.png')

print(f"Probabilidad de compra en menos de {limite_tiempo} min: {prob_30_min:.4f}")
````

````python
## Valor esperado
import numpy as np
import matplotlib.pyplot as plt

# Configuración
n_lanzamientos = 500
probabilidad_exito = 0.5  # Valor Esperado teórico = 0.5

# Simulamos lanzamientos (0 o 1)
resultados = np.random.choice([0, 1], size=n_lanzamientos, p=[1-probabilidad_exito, probabilidad_exito])

# Calculamos el promedio acumulado conforme pasan los lanzamientos
promedio_acumulado = np.cumsum(resultados) / np.arange(1, n_lanzamientos + 1)

# Graficamos
plt.figure(figsize=(10, 5))
plt.plot(promedio_acumulado, label='Promedio Real (Observado)', color='#1f77b4')
plt.axhline(y=probabilidad_exito, color='r', linestyle='--', label=f'Valor Esperado ({probabilidad_exito})')
plt.title('Ley de los Grandes Números: Convergencia al Valor Esperado')
plt.xlabel('Número de Intentos')
plt.ylabel('Valor Promedio')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

#### Medidas de variabilidad

````python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generamos datos sintéticos (n=100 pedidos por sucursal)
np.random.seed(42)
# Ambas tienen una media de 30 minutos, pero diferente dispersión
sucursal_a = np.random.normal(loc=30, scale=2, size=100)  # Desviación pequeña (2)
sucursal_b = np.random.normal(loc=30, scale=8, size=100)  # Desviación grande (8)

def calcular_variabilidad(datos, nombre):
    media = np.mean(datos)
    varianza = np.var(datos, ddof=1) # S² (n-1)
    desviacion = np.std(datos, ddof=1) # S
    cv = (desviacion / media) * 100 # Coeficiente de Variación
    
    print(f"--- {nombre} ---")
    print(f"Media: {media:.2f} min")
    print(f"Varianza (S²): {varianza:.2f}")
    print(f"Desviación Estándar (S): {desviacion:.2f} min")
    print(f"Coeficiente de Variación (CV): {cv:.2f}%\n")
    return media, desviacion

# Ejecutar cálculos
m_a, s_a = calcular_variabilidad(sucursal_a, "SUCURSAL A (Estable)")
m_b, s_b = calcular_variabilidad(sucursal_b, "SUCURSAL B (Inestable)")

# 2. Visualización
plt.figure(figsize=(12, 6))

# Histogramas
plt.hist(sucursal_a, bins=15, alpha=0.5, label='Sucursal A', color='blue')
plt.hist(sucursal_b, bins=15, alpha=0.5, label='Sucursal B', color='orange')

# Líneas de Media y Desviación (Para la Sucursal B que es más visible)
plt.axvline(30, color='red', linestyle='--', label='Media (30 min)')
plt.errorbar(30, 15, xerr=s_b, fmt='o', color='black', label='Dispersión (S) de B')

plt.title("Comparación de Variabilidad en Tiempos de Entrega")
plt.xlabel("Minutos")
plt.ylabel("Cantidad de Pedidos")
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1. GENERACIÓN DE DATOS SINTÉTICOS (Proceso de llenado de botellas)
# Escenario: Queremos llenar 500ml exactos. 
# La precisión es dinero: si llenas de más, pierdes producto; si llenas de menos, hay multas.
np.random.seed(42)
n = 100
maquina_a = np.random.normal(loc=500, scale=1.5, size=n) # Máquina Nueva (S pequeña)
maquina_b = np.random.normal(loc=500, scale=8.0, size=n) # Máquina Descalibrada (S grande)

# 2. CÁLCULO DE MEDIDAS DE VARIABILIDAD
def analizar_variabilidad(datos, nombre):
    media = np.mean(datos)
    varianza = np.var(datos, ddof=1)    # S²: Promedio de distancias al cuadrado
    desviacion = np.std(datos, ddof=1)  # S: Dispersión en unidades reales (ml)
    cv = (desviacion / media) * 100     # CV%: Variabilidad relativa al promedio
    
    return {
        "Máquina": nombre,
        "Media_ml": round(media, 2),
        "Varianza_S2": round(varianza, 2),
        "Desv_Est_S": round(desviacion, 2),
        "CV_Porcentaje": round(cv, 2)
    }

res_a = analizar_variabilidad(maquina_a, "Nueva (A)")
res_b = analizar_variabilidad(maquina_b, "Vieja (B)")

# 3. IMPRESIÓN DE RESULTADOS E INSIGHTS (Todo en la misma ejecución)
df = pd.DataFrame([res_a, res_b])
print("=== TABLA DE CONTROL DE CALIDAD ===")
print(df.to_string(index=False))

print("\n=== INSIGHTS DE NEGOCIO ANALIZADOS ===")
# Insight 1: Sobre la Desviación Estándar (S)
print(f"-> La Máquina A es mucho más confiable: su error típico es de solo {res_a['Desv_Est_S']} ml,")
print(f"   mientras que la B se desvía en promedio {res_b['Desv_Est_S']} ml por botella.")

# Insight 2: Sobre la Varianza (S²)
factor_varianza = round(res_b['Varianza_S2'] / res_a['Varianza_S2'], 1)
print(f"-> ALERTA: La inconsistencia (Varianza) de la Máquina B es {factor_varianza} veces")
print(f"   mayor que la de la Máquina A. Esto indica un proceso fuera de control.")

# Insight 3: Sobre el Coeficiente de Variación (CV)
if res_b['CV_Porcentaje'] > 1.0:
    print(f"-> RECOMENDACIÓN: El CV de la Máquina B ({res_b['CV_Porcentaje']}%) supera el umbral")
    print(f"   de tolerancia del 1%. Se requiere mantenimiento urgente.")

# 4. VISUALIZACIÓN PARA CONFIRMAR LOS INSIGHTS
plt.figure(figsize=(12, 5))

# Histograma: Muestra la "forma" de la variabilidad
plt.subplot(1, 2, 1)
plt.hist(maquina_a, bins=15, alpha=0.7, label=f"Máquina A (S={res_a['Desv_Est_S']})", color='#2ecc71')
plt.hist(maquina_b, bins=15, alpha=0.5, label=f"Máquina B (S={res_b['Desv_Est_S']})", color='#e74c3c')
plt.axvline(500, color='black', linestyle='--', label='Objetivo 500ml')
plt.title('Dispersión de los Datos (Histograma)')
plt.xlabel('Mililitros')
plt.ylabel('Cantidad de Botellas')
plt.legend()

# Boxplot: Muestra qué tan "estirados" están los datos y los valores extremos
plt.subplot(1, 2, 2)
plt.boxplot([maquina_a, maquina_b], labels=['Máquina A', 'Máquina B'], patch_artist=True, 
            boxprops=dict(facecolor='#3498db', alpha=0.6))
plt.title('Rango de Variabilidad (Boxplot)')
plt.ylabel('Mililitros')

plt.tight_layout()
plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 1. DEFINICIÓN DEL ESCENARIO
# Queremos estimar la altura promedio de una población (Valor Real = 170 cm)
valor_real = 170
np.random.seed(42)

# Simulamos 1000 experimentos de estimación (muestreos)
# Estimador A: Insesgado (en promedio acierta, pero es disperso)
estimaciones_a = np.random.normal(loc=170, scale=10, size=1000)

# Estimador B: Sesgado (siempre mide un poco más, pero es muy preciso/junto)
estimaciones_b = np.random.normal(loc=175, scale=2, size=1000)

# 2. CÁLCULO DE SESGO Y ECM
def calcular_metricas(estimaciones, real):
    sesgo = np.mean(estimaciones) - real
    varianza = np.var(estimaciones)
    ecm = np.mean((estimaciones - real)**2)
    # Nota: El ECM también es = Sesgo^2 + Varianza
    return sesgo, ecm

sesgo_a, ecm_a = calcular_metricas(estimaciones_a, valor_real)
sesgo_b, ecm_b = calcular_metricas(estimaciones_b, valor_real)

# 3. IMPRESIÓN DE INSIGHTS
print("-" * 60)
print(f"VALOR REAL A ESTIMAR: {valor_real} cm")
print("-" * 60)
print(f"ESTIMADOR A (Insesgado pero ruidoso):")
print(f"   > Sesgo: {sesgo_a:.2f} (Casi cero, ¡es honesto!)")
print(f"   > ECM:   {ecm_a:.2f} (Error alto por la dispersión)")

print(f"\nESTIMADOR B (Sesgado pero consistente):")
print(f"   > Sesgo: {sesgo_b:.2f} (Sobreestima por 5 cm)")
print(f"   > ECM:   {ecm_b:.2f} (Error total menor que A)")
print("-" * 60)

# --- COMENTARIOS: LOGROS DEL ANÁLISIS ---
# 1. EL DILEMA SESGO-VARIANZA:
#    Logramos ver que un estimador puede ser "perfecto" en promedio (Sesgo 0)
#    pero ser tan variable que el ECM sea alto. A veces preferimos un pequeño
#    sesgo si eso reduce mucho la variabilidad (como el Estimador B).

# 2. EL ECM COMO JUEZ FINAL:
#    El ECM es la métrica de oro porque penaliza tanto el sesgo como la falta 
#    de precisión. En este ejemplo, el Estimador B es "mejor" globalmente.

# 4. VISUALIZACIÓN
plt.figure(figsize=(10, 6))
plt.axvline(valor_real, color='black', linestyle='--', label='Valor Real (170cm)')
plt.hist(estimaciones_a, bins=30, alpha=0.5, label=f'A: Insesgado (ECM: {ecm_a:.1f})', color='blue')
plt.hist(estimaciones_b, bins=30, alpha=0.5, label=f'B: Sesgado (ECM: {ecm_b:.1f})', color='red')

plt.title('Sesgo vs Error Cuadrático Medio (ECM)')
plt.xlabel('Valor Estimado')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
````

#### Estimador de máxima verosimilitud

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# 1. DATOS OBSERVADOS (La Realidad del Pescador)
n_peces = 10     # Total de peces en la red
k_truchas = 7    # Truchas encontradas
# Queremos encontrar 'p' (proporción de truchas en el lago)

# 2. PROBAMOS DIFERENTES HIPÓTESIS (Posibles valores de p)
proporciones_candidatas = np.linspace(0, 1, 100)
verosimilitud = binom.pmf(k_truchas, n_peces, proporciones_candidatas)

# 3. ENCONTRAR EL MÁXIMO (El EMV)
indice_max = np.argmax(verosimilitud)
emv = proporciones_candidatas[indice_max]

# --- PRINTS DE SIGNIFICADO ---
print("-" * 60)
print(f"🐟 RESULTADOS DEL EXPERIMENTO DE PESCA")
print("-" * 60)
print(f"Muestra: {n_peces} peces totales, {k_truchas} son truchas.")
print(f"Estimador de Máxima Verosimilitud (EMV): {emv:.2f}")
print(f"\n¿QUÉ SIGNIFICA ESTE {emv:.2f}?")
print(f"Significa que, de todas las realidades posibles en el lago,")
print(f"la que hace que sacar 7 de 10 truchas sea LO MÁS PROBABLE")
print(f"es que el lago tenga un {emv*100:.0f}% de truchas.")
print("-" * 60)

# --- INSIGHTS ---
# 1. LA "MONTAÑA" DE CREDIBILIDAD: 
#    Si el lago tuviera solo 10% de truchas, la verosimilitud es casi 0 (imposible pescar 7).
#    A medida que nos acercamos al 70%, la probabilidad de observar esos datos sube.

# 2. EL EMV ES INTUITIVO: 
#    En este caso, el EMV coincide con la proporción de la muestra (7/10). 
#    La matemática confirma nuestra intuición: la muestra es el mejor reflejo de la población.

# 4. VISUALIZACIÓN
plt.figure(figsize=(10, 6))
plt.plot(proporciones_candidatas, verosimilitud, color='darkblue', linewidth=3, label='Función de Verosimilitud')
plt.fill_between(proporciones_candidatas, verosimilitud, color='skyblue', alpha=0.4)

# Marcar el punto máximo
plt.axvline(emv, color='red', linestyle='--', label=f'Máximo (EMV = {emv:.2f})')
plt.scatter(emv, np.max(verosimilitud), color='red', s=100, zorder=5)

plt.title('¿Qué porcentaje de truchas hay en el lago?', fontsize=14)
plt.xlabel('Proporción de truchas hipotética (p)')
plt.ylabel('Verosimilitud (Qué tan creíble es)')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

#### Prueba de hipótesis paso a paso

##### Caso de uso Escenario: El equipo de Marketing de una tienda online cree que cambiar el color del botón de "Comprar" de Azul a Verde aumentará la tasa de conversión (ventas). Actualmente, la tasa de conversión es del 12%.


````python
# CONFIGURACION DE DATOS Y SIMULACION

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Definición de parámetros (El "Juicio"):
# H0 (Nula): La conversión es <= 12% (El cambio no sirve).
# Ha (Alternativa): La conversión es > 12% (El botón verde es mejor).
tasa_base = 0.12 
n_muestra = 1000 # Tamaño de usuarios expuestos al botón verde
nivel_significancia = 0.05 # Alfa: Nuestro límite para el Error Tipo I (5%)

# 2. Simulación de resultados reales:
# Generamos datos donde la conversión real del botón verde fue del 15%
np.random.seed(42)
datos_conversion = np.random.binomial(1, 0.15, n_muestra)

# 3. Cálculo de la media observada:
media_observada = np.mean(datos_conversion)
print(f"Conversión obtenida con el botón verde: {media_observada:.2%}")
````

````python
# CALCULO DEL ESTADISTICO Z Y P-VALOR

# 1. Cálculo del Error Estándar:
# Representa la variabilidad esperada si la hipótesis nula fuera cierta.
error_estandar = np.sqrt(tasa_base * (1 - tasa_base) / n_muestra)

# 2. Cálculo del Estadístico Z:
# Mide a cuántas "desviaciones estándar" está nuestra muestra del 12%.
z_score = (media_observada - tasa_base) / error_estandar

# 3. Cálculo del P-Valor:
# Es la probabilidad de obtener este resultado (o uno más extremo) por puro azar.
# Usamos '1 - cdf' porque es una prueba de una sola cola (queremos ver si es MAYOR).
p_valor = 1 - stats.norm.cdf(z_score)

print(f"Estadístico Z: {z_score:.4f}")
print(f"P-Valor: {p_valor:.4f}")
````

````python
# VISUALIZACION DE RESULTADOS
# 1. Crear la curva de la Hipótesis Nula (Distribución Normal Estándar)
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribución H0 (Lo normal)', color='black')

# 2. Marcar el Nivel de Significancia (Alfa = 0.05)
# Si el Z-score cae a la derecha de esta línea roja, rechazamos H0.
z_critico = stats.norm.ppf(1 - nivel_significancia)
plt.axvline(z_critico, color='red', linestyle='--', label=f'Límite Crítico (Alfa 0.05)')
plt.fill_between(x, y, where=(x > z_critico), color='red', alpha=0.2, label='Zona de Rechazo')

# 3. Dibujar nuestro resultado real (Z-Score)
plt.axvline(z_score, color='green', linewidth=3, label=f'Nuestro Resultado (Z={z_score:.2f})')

plt.title('Visualización de la Prueba de Hipótesis')
plt.xlabel('Desviaciones Estándar (Z)')
plt.legend()
plt.show()
````

````python
# TOMA DE DECISION DE NEGOCIO
# 1. Comparación lógica:
if p_valor < nivel_significancia:
    decision = "RECHAZAR H0"
    conclusion = "El botón verde SI aumenta significativamente las ventas."
else:
    decision = "NO RECHAZAR H0"
    conclusion = "No hay pruebas suficientes; el cambio pudo ser suerte."

# 2. Evaluación de Riesgos (Errores):
# Riesgo de Error Tipo I: Cambiar el botón y que las ventas NO suban realmente.
# Riesgo de Error Tipo II: No cambiarlo y perder dinero por ignorar una mejora real.

print(f"Decisión Final: {decision}")
print(f"Conclusión de Negocio: {conclusion}")
````

````python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# =============================================================================
# 1. CONFIGURACIÓN DEL ESCENARIO (EL "BUSINESS CASE")
# =============================================================================
# Queremos validar si cambiar el botón de "Comprar" de Azul a Verde aumenta 
# la conversión del 12% actual. El ticket promedio de venta es de $50 USD.

np.random.seed(42)  # Para que los resultados sean siempre los mismos
n_usuarios = 1000   # Tamaño de la muestra (tráfico enviado al nuevo botón)
tasa_base = 0.12    # Rendimiento histórico del botón azul (H0)
tasa_real_v = 0.15  # Lo que realmente está pasando en el test (simulado)
ticket_medio = 50   # Valor de cada venta

# Generamos los datos: 1 = Compra, 0 = No Compra
datos_verde = np.random.binomial(1, tasa_real_v, n_usuarios)
conversion_observada = np.mean(datos_verde)

# =============================================================================
# 2. CÁLCULO ESTADÍSTICO (MIDIENDO LA INCERTIDUMBRE)
# =============================================================================
# Calculamos el error estándar: qué tanto pueden variar los datos por azar.
error_estandar = np.sqrt(tasa_base * (1 - tasa_base) / n_usuarios)

# Calculamos el Z-Score: cuántas "unidades de sorpresa" se aleja el resultado de lo normal.
z_score = (conversion_observada - tasa_base) / error_estandar

# Calculamos el P-Valor: La probabilidad de que este éxito sea pura suerte.
# Si el p-valor < 0.05, decimos que el resultado es "Significativo".
p_valor = 1 - stats.norm.cdf(z_score)

# =============================================================================
# 3. TRADUCCIÓN TÉCNICA A CONCEPTOS DE GERENCIA
# =============================================================================
# Error Tipo I (Falsa Alarma): Decir que el verde es mejor cuando no lo es. 
# Se controla con el nivel de significancia (alfa = 5%).
confianza_obtenida = (1 - p_valor) * 100

# Error Tipo II (Oportunidad Perdida): No implementar el verde aunque sí sea mejor.
# Esto suele ocurrir cuando la muestra es muy pequeña.

# =============================================================================
# 4. VISUALIZACIÓN DE IMPACTO PARA TOMA DE DECISIONES
# =============================================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
sns.set_style("white")

# --- Gráfico 1: Impacto Financiero Proyectado ---
# Comparamos ingresos estimados por cada 1000 visitas
ingresos_actual = n_usuarios * tasa_base * ticket_medio
ingresos_nuevo = n_usuarios * conversion_observada * ticket_medio

barras = ax1.bar(['Botón Azul (Actual)', 'Botón Verde (Test)'], 
                 [ingresos_actual, ingresos_nuevo], 
                 color=['#bdc3c7', '#27ae60'], edgecolor='black')

ax1.set_title('Ingresos Proyectados por cada 1,000 visitas', fontsize=14, fontweight='bold')
ax1.set_ylabel('Ingresos en USD ($)', fontsize=12)

# Añadimos etiquetas de valor sobre las barras
for bar in barras:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 100, f'${yval:,.0f}', 
             ha='center', va='bottom', fontweight='bold', fontsize=12)

# --- Gráfico 2: Semáforo de Riesgo (Certeza Estadística) ---
# Representamos visualmente qué tan seguros estamos de no cometer un Error Tipo I
ax2.barh(['Nivel de Certeza'], [0.95], color='#ecf0f1', label='Umbral Mínimo (95%)')
ax2.barh(['Nivel de Certeza'], [1 - p_valor], color='#2980b9', height=0.4, label='Certeza Obtenida')

ax2.set_xlim(0.85, 1.0)
ax2.set_title('Validación de Certeza vs. Azar', fontsize=14, fontweight='bold')
ax2.set_xlabel('Probabilidad de Éxito Real (1 - P-valor)')

# Colocamos la leyenda fuera para que no estorbe
ax2.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=True)

# --- TEXTO DE INSIGHTS FINALES ---
# Resumen ejecutivo al pie del gráfico
texto_resumen = (
    f"INSIGHTS PARA GERENCIA:\n"
    f"• El botón verde aumentó la conversión de {tasa_base:.1%} a {conversion_observada:.1%}.\n"
    f"• Esto representa un incremento de ${ingresos_nuevo - ingresos_actual:,.0f} USD en ingresos por cada 1k usuarios.\n"
    f"• La probabilidad de que este resultado sea coincidencia es de apenas {p_valor:.2%}.\n"
    f"• CONCLUSIÓN: Implementar el cambio. El riesgo de error es despreciable."
)

plt.figtext(0.5, 0.02, texto_resumen, ha="center", fontsize=12, 
            bbox={"facecolor":"#f8f9fa", "alpha":0.9, "pad":15, "edgecolor":"#dee2e6"})

plt.tight_layout(rect=[0, 0.15, 1, 0.95])
plt.show()
````

````python

import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# =============================================================================
# 1. DEFINICIÓN DEL PROBLEMA Y DATOS (EXTRACCIÓN DEL ERP)
# =============================================================================
# Hipótesis Nula (H0): La recomendación no cambia la compra (Tasa B <= Tasa A).
# Hipótesis Alternativa (Ha): La recomendación aumenta la compra (Tasa B > Tasa A).

np.random.seed(42)
n_usuarios = 2500 # Tamaño de muestra por cada versión del checkout

# Simulamos comportamiento: el botón de recomendación sube la conversión del 10% al 12%
tasa_base = 0.10
datos_control = np.random.binomial(1, tasa_base, n_usuarios)    # Grupo A (Sin recomendación)
datos_test = np.random.binomial(1, 0.125, n_usuarios)           # Grupo B (Con recomendación)

# =============================================================================
# 2. CÁLCULO DE MÉTRICAS CLAVE
# =============================================================================
exitos_a, exitos_b = datos_control.sum(), datos_test.sum()
tasa_a, tasa_b = exitos_a / n_usuarios, exitos_b / n_usuarios

# Estadístico Z para Proporciones (Mide la fuerza de la evidencia)
p_combinada = (exitos_a + exitos_b) / (2 * n_usuarios)
error_estandar = np.sqrt(p_combinada * (1 - p_combinada) * (2 / n_usuarios))
z_stat = (tasa_b - tasa_a) / error_estandar

# P-Valor: Probabilidad de que el éxito de 'B' sea por pura casualidad
p_valor = 1 - stats.norm.cdf(z_stat)

# =============================================================================
# 3. VISUALIZACIÓN DE RESULTADOS PARA PRESENTACIÓN
# =============================================================================
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
sns.set_style("whitegrid")

# Gráfico A: Comparación de Rendimiento (Métrica de Negocio)
colores = ['#95a5a6', '#3498db'] # Gris (Control) vs Azul (Nuevo)
barras = ax1.bar(['Checkout Estándar', 'Checkout con Recs.'], [tasa_a, tasa_b], color=colores)
ax1.set_title('Tasa de Conversión: ¿Funcionan las Recomendaciones?', fontsize=14, fontweight='bold')
ax1.set_ylim(0, 0.15)
for b in barras:
    ax1.text(b.get_x() + b.get_width()/2, b.get_height() + 0.005, f'{b.get_height():.2%}', ha='center', fontweight='bold')

# Gráfico B: Curva de Probabilidad (Validación Estadística)
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)
ax2.plot(x, y, color='black')
z_critico = stats.norm.ppf(0.95) # Límite del 5% de riesgo (Alfa)
ax2.fill_between(x, y, where=(x >= z_critico), color='red', alpha=0.3, label='Zona de Rechazo (Error Tipo I)')
ax2.axvline(z_stat, color='blue', linestyle='--', linewidth=2, label=f'Resultado Observado (Z={z_stat:.2f})')
ax2.set_title('Evidencia vs. Azar', fontsize=14, fontweight='bold')
ax2.legend(loc='upper left', bbox_to_anchor=(0, -0.1))

# --- INSIGHTS FINALES ---
texto_final = (
    f"RESUMEN PARA GERENCIA:\n"
    f"• La tasa de conversión subió de {tasa_a:.1%} a {tasa_b:.1%}.\n"
    f"• El P-valor es {p_valor:.4f} (Menor al 0.05 permitido).\n"
    f"• CONCLUSIÓN: El resultado es significativo. Las recomendaciones NO distraen,\n"
    f"  sino que impulsan la venta cruzada. Implementar en producción."
)
plt.figtext(0.5, 0.02, texto_final, ha="center", fontsize=12, bbox={"facecolor":"#f0f7fb", "alpha":0.9, "pad":10})
plt.tight_layout(rect=[0, 0.15, 1, 0.95])
plt.show()
````

#### ANOVA

````python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# =============================================================================
# CASO PRÁCTICO: ¿QUÉ RUTINA DE EJERCICIO QUEMA MÁS CALORÍAS?
# Contexto: Un gimnasio prueba 3 rutinas (Cardio, Pesas, Yoga) en 15 personas.
# Objetivo: Saber si el tipo de rutina afecta el promedio de calorías quemadas.
# =============================================================================

# 1. DATOS SIMPLES (Diccionario de Python)
datos = {
    'Rutina': ['Cardio']*5 + ['Pesas']*5 + ['Yoga']*5,
    'Calorias': [350, 400, 380, 420, 390,  # Cardio (Media alta)
                 300, 320, 310, 290, 330,  # Pesas  (Media media)
                 150, 180, 160, 200, 170]  # Yoga   (Media baja)
}

df = pd.DataFrame(datos)

# 2. ANÁLISIS VISUAL BÁSICO
# El primer paso es ver si las medias se ven distintas "a ojo".
print("--- PROMEDIOS POR GRUPO ---")
print(df.groupby('Rutina')['Calorias'].mean())

df.boxplot(column='Calorias', by='Rutina')
plt.title('Calorías quemadas según la Rutina')
plt.suptitle('') # Limpiar título automático
plt.show()

# [Image of boxplot for ANOVA showing three treatment groups with different means]

# 3. EL CORAZÓN DEL ANÁLISIS: ANOVA de una vía
# Usamos la función f_oneway de la librería scipy.stats.
# H0 (Hipótesis Nula): Todas las rutinas queman lo mismo.
# H1 (Hipótesis Alterna): Al menos una rutina es diferente.

grupo_cardio = df[df['Rutina'] == 'Cardio']['Calorias']
grupo_pesas = df[df['Rutina'] == 'Pesas']['Calorias']
grupo_yoga = df[df['Rutina'] == 'Yoga']['Calorias']

f_stat, p_valor = stats.f_oneway(grupo_cardio, grupo_pesas, grupo_yoga)

# 4. INTERPRETACIÓN DE RESULTADOS (INSIGHTS)
print("\n--- RESULTADOS ESTADÍSTICOS ---")
print(f"Estadístico F: {f_stat:.2f}")
print(f"p-value: {p_valor:.6f}")

# [Image of F-distribution curve showing the rejection region for ANOVA]

print("\n--- CONCLUSIÓN ---")
if p_valor < 0.05:
    # INSIGHT: Si el p-valor es pequeño, las diferencias NO son casualidad.
    print("INSIGHT: El tipo de rutina SÍ influye en las calorías quemadas.")
    print("ACCIÓN: El gimnasio debería promocionar Cardio para pérdida de peso rápida.")
else:
    # INSIGHT: Si el p-valor es grande, cualquier diferencia es puro azar.
    print("INSIGHT: No hay diferencia significativa. Todas queman parecido.")
    print("ACCIÓN: El cliente puede elegir la que más le guste sin culpa.")

# 5. EXTRA: Diferencias individuales (Sin librerías complejas)
# Si queremos ver quién es diferente de quién sin usar Post-Hoc avanzados, 
# podemos comparar las medias restándolas:
diff = grupo_cardio.mean() - grupo_yoga.mean()
print(f"\nNota adicional: Cardio quema en promedio {diff} calorías más que Yoga.")
````

#### Caso práctico Regresion lineal multiple



````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Generación de datos sintéticos
np.random.seed(42)
n_puntos = 100
marketing = np.random.normal(500, 100, n_puntos)    # Inversión diaria
trafico = (marketing * 1.2) + np.random.normal(100, 30, n_puntos) # El tráfico depende del marketing
descuento = np.random.uniform(0.05, 0.30, n_puntos) # Porcentaje de descuento

# Variable dependiente: Ventas (con un poco de ruido aleatorio)
ventas = (2.5 * marketing) + (0.5 * trafico) + (1000 * descuento) + np.random.normal(0, 50, n_puntos)

df = pd.DataFrame({
    'Marketing': marketing,
    'Trafico': trafico,
    'Descuento': descuento,
    'Ventas': ventas
})

print("Dataset de E-commerce listo:")
print(df.head())
````

````python
# Verificacion de supuestos Linealidad y Multicolinealidad. Se debe verificar que existe una relacion lineal
# y que las variables independientes no estan excesivamente correlacionadas entre si (Multicolinealidad)

# 1. Visualizar la relación lineal de cada variable contra Ventas
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(['Marketing', 'Trafico', 'Descuento']):
    sns.scatterplot(data=df, x=col, y='Ventas', ax=axes[i])
    axes[i].set_title(f'{col} vs Ventas')
plt.show()

# 2. Calcular el Factor de Inflación de la Varianza (VIF)
# Si el VIF > 10, hay mucha multicolinealidad y el modelo podría ser inestable [cite: 44]
X_vif = df[['Marketing', 'Trafico', 'Descuento']]
vif_data = pd.DataFrame()
vif_data["Variable"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(len(X_vif.columns))]

print("Análisis de Multicolinealidad (VIF):")
print(vif_data)
# Insight: Un VIF bajo asegura que cada variable aporta información única al modelo[cite: 40, 41].
````

````python
# Construccion de Modelo y calculo de coeficientes 
# Definir variables independientes (X) y dependiente (y)
X = df[['Marketing', 'Trafico', 'Descuento']]
y = df['Ventas']

# Añadir una constante para el Intercepto (Paso necesario en statsmodels)
X_with_const = sm.add_constant(X)

# Ajustar el modelo
modelo_rlm = sm.OLS(y, X_with_const).fit()

# Mostrar coeficientes
print("Resultados de los Coeficientes:")
print(modelo_rlm.params)

# Insight: El coeficiente de 'Marketing' indica cuánto aumentan las ventas por cada 
# unidad extra invertida, manteniendo el tráfico y descuento constantes[cite: 31, 50].
````

````python
# Obtener el R2
# Se determina el porcentaje de variacion de las ventas 


# Obtener R-cuadrado y R-cuadrado ajustado
r2 = modelo_rlm.rsquared
r2_adj = modelo_rlm.rsquared_adj
p_f_stat = modelo_rlm.f_pvalue # Prueba F para el modelo global

print(f"Coeficiente de Determinación (R²): {r2:.4f}")
print(f"R² Ajustado: {r2_adj:.4f}")
print(f"Valor p del modelo (Prueba F): {p_f_stat:.4e}")

# Insight: Si el R² es 0.9337, el 93.37% de las ventas se explican por estas 3 variables[cite: 15, 56].
# Si el Valor p < 0.05, el modelo es significativo y se puede generalizar[cite: 23, 53].
````

````python
## Analisis de residuos o errores. Se evalua el error del análisis para ver si las predicciones
## se desvian mucho de la realidad 

# Calcular los residuos (Ventas reales - Ventas predichas)
residuos = modelo_rlm.resid

plt.figure(figsize=(10, 5))

# Histograma de residuos para verificar Normalidad
plt.subplot(1, 2, 1)
sns.histplot(residuos, kde=True, color='purple')
plt.title('Distribución de los Errores')

# Gráfico de dispersión de residuos para verificar Homocedasticidad
plt.subplot(1, 2, 2)
plt.scatter(modelo_rlm.predict(), residuos)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuos vs Predicciones')
plt.xlabel('Ventas Predichas')
plt.ylabel('Error')

plt.tight_layout()
plt.show()

# Insight: Si los residuos están dispersos de forma aleatoria alrededor de cero, 
# el modelo es confiable y no tiene sesgos importantes.
````
